```python
import re
```

```python
def cleantext(text):
    text =text.lower()

    #remove url
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    #remove HTML tag
    text = re.sub(r'<.*?>', '', text)

    #remove digits
    text = re.sub(r'\d', '', text)
    
    #remove punction
    text = re.sub(r'[^\w\s]', '', text)
    
    #remove \n
    text = re.sub(r'\n', ' ', text)
    return text
```

```python
news['text'] = news['text'].apply(cleantext)
```

#this makes easy for ML model to understand

```python
x=news['text']
y=news['lable']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size = 0.3) 
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorfunc = TfidfVectorizer()
```

```python
xv_train = vectorfunc.fit_transform(x_train)
```

```python
xv_test = vectorfunc.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

#Logistic regression

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr =LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
DCT = DecisionTreeClassifier()
```

```python
DCT.fit(xv_train, y_train)
```

```python
pred_dct = DCT.predict(xv_test)
```

```python
DCT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dct))
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
gbc=GradientBoostingClassifier()
```

```python
def output_lable(n):
    if n==0:
        return "It is False, Misleading and Fake News"
    elif n==1:
        return "It is a Genuine News, Real News"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]} #defining dictionary
    new_test = pd.DataFrame(testing_news)
    new_test["text"] = new_test["text"].apply(cleantext)
    new_x_test = new_test["text"]
    new_xv_test = vectorfunc.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    #pred_dtc = dtc.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    return "\n\nLR Prediction:{} \nGBC Prediction:{} \nRFC Prediction:{}".format(
        output_lable(pred_lr[0]), output_lable(pred_gbc[0]), output_lable(pred_rfc[0])
    )
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
